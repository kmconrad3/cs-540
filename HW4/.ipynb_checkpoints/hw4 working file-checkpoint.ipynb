Data structure needed:
1. the list of clusters, and also which data points are in these clusters 
The clusters are indexed by the time it gets created. In particular, the cluster created at iteration t 
is indexed by n-1+t. 

 1) list of clusters: array of cluster indices
 2) list of data points indices for each cluster in the current list of clusters.

2. the distance between the data points
implement this by keeping a distance matrix for all data points

3. enumerate all the pairs of clusters in the current list, and compute the pairwise distances between them
implement it by computing a distance matrix between the current clusters

4. the array Z as in the outline in the pdf

Pseudo-code:
1. compute the distance matrix for the data points. initialize list of clusters, and for each cluster 
the list of points in this cluster. initialize Z.


2. For t = 1,2,…,n-1:  
 1) create a distance matrix between the current list of clusters. 
compute the distances between clusters using the information about the current clusters and distance matrix
between points.
 2) find the closet pair of clusters. note that you may need tie-breaking rule as in the pdf

3) merge the two clusters: record this merge in the output matrix Z; update the list of clusters, and also 
update the list of lists of data points.

say n =5. in first iteration, we merge cluster 1 and 2
before 1st iteration: list of clusters [0, 1, 2, 3, 4], and for each cluster we have a list:
        [0], [1], [2], [3], [4]
        
after 1st iter merging cluster 1 and 2: list of clusters [0, 3, 4, 5], and for each cluster we have a list: 
        [0], [3], [4], [1,2]

```python
def load_data(filepath):
    with open(filepath, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        poke_list=[]
        for row in reader:
            poke_list.append({'HP': row['HP'], 
                          'Attack': row['Attack'], 
                          'Defense': row['Defense'], 
                          'Sp. Atk': row['Sp. Atk'], 
                          'Sp. Def': row['Sp. Def'], 
                          'Speed': row['Speed']})
    return poke_list


def calc_features(row):
    raw = np.fromiter(row.values(), dtype=np.int64)
    return np.array((raw[1], raw[3], raw[5], raw[2], raw[4], raw[0]))


def hac(dataset):
    Z = []
    clusters = []
    n = len(dataset)
    
    for i in range(n):

        clusters.append([i,tuple([i])])
    
    distances = distance_calc2(dataset)
    
    for i in range(len(distances)-1):
        
        next_cluster = calc_cluster(clusters,distances)
        idx_i = next_cluster[0]
        idx_j = next_cluster[1]
        dist = int(next_cluster[2])
        
        Z, clusters = merge(Z, dataset, clusters, i, idx_i, idx_j, dist)
             
    return np.array(Z).astype("float")


def imshow_hac(Z):
    #plt.figure()
    dn = hierarchy.dendrogram(Z)
    plt.show()

    
def merge(Z, dataset, clusters, n, idx_i, idx_j, dist):
    
    for i in clusters:
        idx = i[0]
        if idx == idx_i:
            cluster_i = i
        if idx == idx_j:
            cluster_j = i
    
    m = len(dataset)
    
    cluster1 = list(cluster_i[1])
    cluster2 = list(cluster_j[1])
    
    cluster1.extend(cluster2)
    
    minn=0
    maxx=0
    if idx_i <idx_j:
        minn= idx_i
    else:
        minn= idx_j
    
    if idx_i >idx_j:
        maxx= idx_i
    else:
        maxx=idx_j
        
    Z.append([minn, maxx, dist, len(cluster1)]) 
    
    nc = [m+n, tuple(cluster1)]
    clusters.remove(cluster_i)
    clusters.remove(cluster_j)
    clusters.append(nc)
    return Z, clusters


def calc_cluster(clusters,distances):
    
    answer_i = -1
    answer_j = -1
    mini = 10**9
    for i in clusters:
        
        idx_i = i[0]
        cli_i=i[1]
        
        for j in clusters:
            idx_j = j[0]
            cli_j = j[1]
             
            if idx_i == idx_j:
                continue
             
            maxi = -1
            psuedo_i = -1
            psuedo_j = -1
            for k in cli_i:
                for l in cli_j:
                    if distances[k][l] > maxi:
                        maxi = float(distances[k,l])
                         
                        psuedo_i = k
                        psuedo_j = l
             
            if mini > maxi and maxi >= 0:
                mini = maxi
                
                answer_i = idx_i
                answer_j = idx_j
                        
    return [answer_i,answer_j,mini]         


def distance_calc2(dataset): #distance matrix
    distances = np.zeros((len(dataset),len(dataset)))
    for i in range(len(dataset)):
        for j in range(len(dataset)):
                
                dist = np.linalg.norm(dataset[i] - dataset[j])
                distances[i][j] = dist
    return distances



data = load_data("Pokemon.csv")[:30]
X = []
for i in data:
    X.append(calc_features(i))
X = hac(X)
print(X)
imshow_hac(X)
```
